````python
import os

def get_id(file_name):
    # On récupère l'identifiant de l'image
    elem1, elem2 = file_name.split(".")[:2]
    if len(elem1) == 1:
        elem1 = '0' + elem1 + '0'
    else:
        elem1 += '0'
    if len(elem2) == 1:
        elem2 = '00' + elem2
    elif len(elem2) == 2:
        elem2 = '0' + elem2
    return elem1 + elem2
````

````python

fichier_data = "/Users/rolly/Documents/10-19_Université_et_scolarité/paleo_comput/data"
data = []

for root, dirs, files in os.walk(fichier_data):
    for file in files:
        if file.endswith(".jpg"):
            # On récupère la classe
            classe = root.split("/")[-2] if len(root.split("/")) == 9 else root.split("/")[-1]
            
            # On récupère le nom et l'identifiant de l'image
            nom_image = file.split(".")[-2].strip()
            try:
                id_image = get_id(file)
            except ValueError:
                print(f"Erreur lors du traitement du fichier : {file}")
                continue
            
            # On ajoute les données dans une liste
            data.append({
                "chemin" : os.path.join(root,file),
                "classe": classe,
                "nom_image": nom_image,
                "id": id_image
            })

# Affichage des données collectées
print(data)
````
